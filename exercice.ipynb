Partie Bonus

Créez une fonction analyser_texte(texte) qui prend une chaîne de caractères en entrée et retourne un dictionnaire contenant les informations suivantes :

Nombre total de mots
Nombre de caractères (espaces inclus)
Nombre de caractères (espaces exclus)
Mot le plus long
Longueur moyenne des mots

Testez cette fonction avec le texte suivant : "Python est un langage de programmation puissant et facile à apprendre. Il dispose de structures de données de haut niveau efficaces et d'une approche simple mais efficace de la programmation orientée objet."

```python


dico = "Python est un langage de programmation puissant et facile à apprendre. Il dispose de structures de données de haut niveau efficaces et d'une approche simple mais efficace de la programmation orientée objet."

#Nombre total de mots
print(f"voici le nombre de mots dans la chaîne =  {len(dico.split())}")

#Nombre total de caractères
print(f"voici la longueur de la chaîne =  {len(dico)}")

#Nombre total de caractères sans les espaces
compteur = 0
for i in dico :
    if i != " " :
        compteur += 1
print(f"voici le nombre de caractères sans les espaces =  {compteur}")

total = 0
mots = dico.split()

#Mot le plus long
mot_le_plus_long = max(mots, key=len)
print(f"voici le mot le plus long dans la chaîne =  {mot_le_plus_long}")


#moyenne de la longueur des mots
for i in mots :
        total += len(mots)

moyenne = total / len(mots)
print(f"voici la longueur moyenne des mots dans la chaîne =  {moyenne}")
```
